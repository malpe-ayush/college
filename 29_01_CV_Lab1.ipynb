```python
nature=cv2.imread("/content/free-nature-images.jpg")
```

```python
cv2_imshow(nature)
```

```python
img1=cv2.cvtColor(nature, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img1)
```

```python
img1
```

```python
scale_percent = 60 # percent of the original size
width = int(img1.shape[1] * scale_percent / 100)
height = int(img1.shape[0] * scale_percent / 100)
dimension = (width, height)
resized_image = cv2.resize(img1, dimension, interpolation = cv2.INTER_AREA)
print('Resized Dimensions : ',resized_image.shape)
cv2_imshow(resized_image)
```

```python
image = cv2.rotate(img1, cv2.ROTATE_90_CLOCKWISE)
```

```python
cv2_imshow(image)
```

```python
M = np.float32([
	[1, 0, -30],
	[0, 1, 12]
])

shifted = cv2.warpAffine(resized_image, M, (resized_image.shape[1], resized_image.shape[0]))
cv2_imshow(shifted)
```

```python
rows, cols = resized_image.shape
N = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(resized_image, N, (int(cols*1.5), int(rows*1.5)))
cv2_imshow(sheared_img)
```

```python
img1 = cv2.imread('/content/1bit1.png')
cv2_imshow(img1)
img2 = cv2.imread('/content/2bit2.png')
cv2_imshow(img2)
```

```python
dest_and = cv2.bitwise_and(img2, img1, mask = None)
cv2_imshow(dest_and)
```

```python
dest_or = cv2.bitwise_or(img2, img1, mask = None)
cv2_imshow(dest_or)
```

```python
dest_xor = cv2.bitwise_xor(img1, img2, mask = None)
cv2_imshow(dest_xor)
```

```python
dest_not1 = cv2.bitwise_not(img1, mask = None)
cv2_imshow(dest_not1)
```

```python
dest_not2 = cv2.bitwise_not(img2, mask = None)
cv2_imshow(dest_not2)
```
