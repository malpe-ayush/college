```python
# Read the image (Fig 2)
img = cv2.imread(r"/content/image_4.png", 0)

# Binarize the image
binr = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Define the kernel
kernel = np.ones((2,2), np.uint8)

# Closing operation to fill gaps
closing = cv2.morphologyEx(binr, cv2.MORPH_CLOSE, kernel, iterations=2)

# Opening operation to remove noise
opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel, iterations=2)

# Display the original, closed, and opened images side by side
plt.subplot(1, 3, 1)
plt.imshow(binr, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2)
plt.imshow(closing, cmap='gray')
plt.title('After Closing')

plt.subplot(1, 3, 3)
plt.imshow(opening, cmap='gray')
plt.title('After Opening')

# Adjust layout for better display
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Read the image (Fig 3)
img2 = cv2.imread(r"/content/image_5.png", cv2.IMREAD_GRAYSCALE)

# Define the kernel for dilation and erosion
kernel = np.ones((3, 3), np.uint8)

# Grayscale Dilation
dilation = cv2.dilate(img2, kernel, iterations=1)

# Grayscale Erosion
erosion = cv2.erode(img2, kernel, iterations=1)

# Morphological Gradient
gradient = cv2.absdiff(dilation, erosion)

# Display the original, dilation, erosion, and gradient images
plt.subplot(1, 4, 1)
plt.imshow(img2, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 4, 2)
plt.imshow(dilation, cmap='gray')
plt.title('Dilation')

plt.subplot(1, 4, 3)
plt.imshow(erosion, cmap='gray')
plt.title('Erosion')

plt.subplot(1, 4, 4)
plt.imshow(gradient, cmap='gray')
plt.title('Morphological Gradient')

# Adjust layout for better display
plt.tight_layout()

# Show the plot
plt.show()

```
